```python
# Replacing Chinese brand names with English names
list_source =  ['华为','小米','三星','vivo','OPPO','魅族','酷派','乐视','联想','HTC']
list_dest = ['Huawei','Xiaomi','Samsung','vivo','OPPO','Meizu','Coolpad','LeEco','Lenovo','HTC']
phone_brand_df['Brand'] = phone_brand_df['Brand'].replace(list_source, list_dest)
```

### Merging Data frames

```python
# Events and gender_age_train
device_event_df = pd.merge(df_events_req_states,gender_a_t_df,on='device_id',how='left')
device_event_df.fillna(0,inplace=True)
device_event_df
```

```python
# Events and phone_brand_device_model
device_event_df = pd.merge(device_event_df,phone_brand_df,on='device_id',how='left')
device_event_df.fillna(0,inplace=True)
device_event_df
```
